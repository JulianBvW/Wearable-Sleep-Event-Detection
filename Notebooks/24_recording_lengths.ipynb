```python
from wearsed.dataset.WearSEDDataset import WearSEDDataset
import matplotlib.pyplot as plt
from tqdm import tqdm
import pandas as pd
```

```python
dataset = WearSEDDataset(mesaid_path='../wearsed/dataset/data_ids/', signals_to_read=[], return_recording=True)

lens = []
for i in tqdm(range(len(dataset))):
    hy = dataset[i].hypnogram
    lens.append(len(hy))
lens = pd.Series(sorted(lens))
```

```python
(lens/60/60).plot(grid=True, ylabel='Hours')
(lens/60/60).describe()
```
