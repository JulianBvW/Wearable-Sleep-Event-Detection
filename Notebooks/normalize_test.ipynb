```python
from wearsed.dataset.Recording import Recording

import pandas as pd
import matplotlib.pyplot as plt

import torch
import torchvision.transforms.functional as TF
```

```python
r1 = Recording(1)

# Original Values
spo2 = r1.psg['SpO2']

# Normalization
spo2_t = torch.Tensor(spo2)

mask = ~spo2_t.isnan()
mean = spo2_t[mask].mean()
std = spo2_t[mask].std()

spo2_t_norm = (spo2_t - mean) / std

# Custom Normalization
spo2_t_norm_custom = (spo2_t - 50) / 66
```

```python
fig, axs = plt.subplots(3, 1, figsize=(20, 3*3))
spo2.plot(ax=axs[0])
pd.Series(spo2_t_norm).plot(ax=axs[1])
pd.Series(spo2_t_norm_custom).plot(ax=axs[2])
plt.show()
```
