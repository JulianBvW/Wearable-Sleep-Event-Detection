```python
from wearsed.training.metric import correct
import matplotlib.pyplot as plt
import pandas as pd
```

```python
output = pd.read_csv('../wearsed/training/baseline_conv/output/new_baseline_BN/test_preds_epoch_31.csv')
y_true, y_pred = output['targets'], output['predictions']
_, axs = plt.subplots(5, 1, figsize=(20, 15))
y_true[:10000].plot(ax=axs[0], color='red')
y_pred[:10000].plot(ax=axs[1])
axs[1].plot([0,10000-1], [0.4]*2, color='green', linestyle='dashed', marker='o')
((y_pred[:10000] > 0.4)*1).plot(ax=axs[2])
pd.Series(correct(y_pred[:10000] > 0.4)*1).plot(ax=axs[3])
y_true[:10000].plot(ax=axs[4], color='red')
```
