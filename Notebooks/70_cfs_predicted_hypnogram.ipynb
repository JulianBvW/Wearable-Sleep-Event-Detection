```python
from wearsed.dataset.CFSDataset import CFSDataset
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
ds = CFSDataset(cfs_id_path='../wearsed/dataset/data_ids/')
```

```python
lens_hypnogram, lens_spo2, lens_ppg, lens_targets = [], [], [], []
for cfs_id in tqdm(range(len(ds))):
    (hypnogram, spo2, ppg), event_or_not = ds[cfs_id]
    lens_hypnogram.append(len(hypnogram))
    lens_spo2.append(len(spo2))
    lens_ppg.append(len(ppg))
    lens_targets.append(len(event_or_not))
```

```python
plt.scatter(lens_hypnogram, lens_spo2, label='Hypnogram vs SPO2')
```

```python
plt.scatter(lens_hypnogram, lens_ppg, label='Hypnogram vs PPG')
```

```python
plt.scatter(lens_hypnogram, lens_targets, label='Hypnogram vs Targets')
plt.xlabel('Length of Hypnogram')
plt.ylabel('Length of Targets')
```
