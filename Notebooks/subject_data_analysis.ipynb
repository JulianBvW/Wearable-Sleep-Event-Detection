```python
#%cd ..

from wearsed.dataset.WearSEDDataset import WearSEDDataset
from collections import Counter
import pandas as pd
```

```python
dataset = WearSEDDataset()
```

```python
subject_infos = {
    'age':         {i: 0 for i in range(-1, 100)}, 
    'bmi':         {i: 0 for i in range(-1, 100)},
    'sex':         [],
    'race':        [],
    'cur_smoker':  [],
    'ever_smoked': []
}
```

```python
def int_or_NaN(n):
    try:
        n = int(n)
        if n >= 100:
            print('Large')
        return n
    except:
        return -1
```

```python
try:
    for idx, recording in enumerate(dataset):
        subject_info = recording.get_subject_info()
        subject_infos['age'][int_or_NaN(subject_info['age'])] += 1
        subject_infos['bmi'][int_or_NaN(subject_info['bmi'])] += 1
        subject_infos['sex'].append(subject_info['sex'])
        subject_infos['race'].append(subject_info['race'])
        subject_infos['cur_smoker'].append(subject_info['cur_smoker'])
        subject_infos['ever_smoked'].append(subject_info['ever_smoked'])
except:
    pass
```

```python
pd.Series(subject_infos['age']).plot()
```

```python
pd.Series(subject_infos['bmi']).plot()
```

```python
print(Counter(subject_infos['sex']))
print(Counter(subject_infos['race']))
print(Counter(subject_infos['cur_smoker']))
print(Counter(subject_infos['ever_smoked']))
```
