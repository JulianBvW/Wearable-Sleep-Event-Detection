```python
from wearsed.dataset.CFSDataset import CFSDataset
import matplotlib.pyplot as plt
```

```python
ds = CFSDataset(cfs_id_path='../wearsed/dataset/data_ids/')
```

```python
len(ds)
```

```python
ds[0]
```

```python
i = 30
(hypnogram, spo2, ppg), cfs_id = ds[i]

_, axs = plt.subplots(3, 1, figsize=(20, 2*3))
axs[0].set_title(f'CFS ID: {cfs_id}')

axs[0].plot(hypnogram)
axs[1].plot(spo2[:5*60])
axs[2].plot(ppg[:5*60*128])

plt.tight_layout()
plt.show()
```
