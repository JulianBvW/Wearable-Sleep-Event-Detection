```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
ahi_scs_nsrr       = pd.read_csv('../wearsed/dataset/data_ids/mesa_ids_nsrr.csv')
ahi_scs_somnolyzer = pd.read_csv('../wearsed/dataset/data_ids/mesa_ids_somnolyzer.csv')

ahi_scs_nsrr       = Counter([int(sc) for sc in ahi_scs_nsrr['ahi_severity_class'].values])
ahi_scs_somnolyzer = Counter([int(sc) for sc in ahi_scs_somnolyzer['ahi_severity_class'].values])

ahi_scs_nsrr       = {i: ahi_scs_nsrr[i] for i in range(4)}
ahi_scs_somnolyzer = {i: ahi_scs_somnolyzer[i] for i in range(4)}
```

```python
severity_classes = ['normal', 'mild', 'moderate', 'severe']
counts = {
    'NSRR': [ahi_scs_nsrr[i] for i in range(4)],
    'SOMNOLYZER': [ahi_scs_somnolyzer[i] for i in range(4)]
}

x = np.arange(len(severity_classes))
width = 0.3
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for scoring, values in counts.items():
    offset = width * multiplier + 0.15
    rects = ax.bar(x + offset, values, width, label=scoring)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_title('Number of patients for each severity class')
ax.set_xticks(x + width, severity_classes)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 900)

plt.show()
```
